```python
df = pd.read_csv('/Users/yabebal/My-Works/Business/tsegi/tsegi_abyssinya_bank_transfers_1.csv',skiprows=4) #,sheet_name=None)
```

```python
pd.set_option('display.max_rows', None)
df
```

```python
def func(x):
    y = x.split('From')[0]
    for t in ['Credit','Transfer no chq','Transfer']:
        if t in y:
            yy = y.split(t)
            if len(yy)>1:
                z = ''.join(xx for xx in yy[1] if not xx.islower())
                return ' '.join(z.split(' ')[0:3]) 
            else:
                z = ''.join(xx for xx in yy[0] if not xx.islower())
                return ' '.join(z.split(' ')[0:3]) 
        
    return pd.NA
        
df['Name'] = df.Description.map(lambda x: func(x))
df
```

```python

dff = df.dropna(axis=0,subset=['Name'])
dff.shape
```

```python
f1 = lambda x: ', '.join(x)
def f2(x):
    y=0
    z = ''
    for i in x:
        y+=i
        z+=str(i)+', '
    return str(len(x))+', '+str(y)
    
cols = ['Date','Amount (ETB)','Name']

dfg = dff[cols].groupby('Name').agg({'Amount (ETB)':f2, 'Date':f1})
dfg
```
